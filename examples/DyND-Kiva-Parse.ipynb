# Traversing Kiva Loans with DyND

We have a 1.5 GB JSON file of all Kiva loans from XXX to XXX

We use describe the format of the data with datashape.
We use DyND to efficiently slice into the data to collect variables of interest.

Once we have this data the we leverage Pandas and Bokeh to provide analytics.

```python
from dynd import nd
```

```python
with open('loans.datashape') as f:
    ds = f.read()
print ds
```

```python
x = nd.parse_json(ds, nd.memmap('loans_data.json'),
                  ectx=nd.eval_context(default_errmode='none'))
```

```python
from pandas import DataFrame
```

While DyND can deal with variable length strings, NumPy can't

Here we need to switch down to Python string objects to build a DataFrame.  Fortunately this is after the slow parsing and slicing steps.

```python
df = DataFrame({'Country':    nd.as_py(x.location.country), 
                'Delinquent': nd.as_py(x.delinquent)})
```

```python
df.groupby('Country')[['Delinquent']].mean().sort('Delinquent', ascending=False)
```
